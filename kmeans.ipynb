```python
#import the all libaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
```

```python
#read the data
data=pd.read_csv(r"C:\Users\Somay\Documents\iris.csv")
print(data)
```

```python
#to measure the central tendency
data.describe()
```

```python
#to check null value in datasets
data.isnull().sum()
```

```python
#define pred and target
x=data.iloc[:,:2].values
y=data.iloc[:,1].values
```

```python
# visulalize the data
plt.scatter(x[:,0],x[:,1],c=y,cmap='gist_rainbow')
plt.xlabel('SepalLengthCm',fontsize=18)
plt.ylabel('SepalWidthCm',fontsize=18)
```

```python
# fit the model
km=KMeans(n_clusters=3,n_jobs=4,random_state=21)
km.fit(x)
```

```python
# identify the centre points
centers=km.cluster_centers_
print(centers)
```

```python
# finally clusters results
new_labels=km.labels_
```

```python
#plot the cluster graph
fig,axes=plt.subplots(1,2,figsize=(14,7))
axes[0].scatter(x[:,0],x[:,1],c=y,cmap='gist_rainbow',edgecolor='k',s=145)
axes[1].scatter(x[:,0],x[:,1],c=new_labels,cmap='jet',edgecolor='k',s=145)
axes[0].set_xlabel('SepalLengthCm',fontsize=18)
axes[0].set_ylabel('SepalWidthCm',fontsize=18)
axes[1].set_xlabel('SepalLengthCm',fontsize=18)
axes[1].set_ylabel('SepalWidthCm',fontsize=18)
axes[0].tick_params(direction='in',length=10,width=5,color='k',labelsize=18)
axes[1].tick_params(direction='in',length=10,width=5,color='k',labelsize=18)
axes[0].set_title('Actual',fontsize=16)
axes[1].set_title('Predicted',fontsize=16)
```
